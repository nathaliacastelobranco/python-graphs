```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
sns.set_context("notebook")
```

```python
data = {
    'Frequência' : [5.61, 5.65, 5.77, 6.01, 6.40, 6.97, 7.73, 8.75],
    'Massa Modal 01': [8385.67, 8440.74, 8645.59, 8552.16, 7670.82, 5332.95, 3659.60, 3151.19],
    'Massa Modal 02': [8385.67, 8440.74, 8645.59, 8552.16, 7670.82, 5332.95, 3659.60, 3151.19],
}
```

```python
df = pd.DataFrame(data)
```

```python
color_map = ['orange','pink']
```

```python
ax = plt.stackplot(df['Frequência'], df['Massa Modal (kg)'], df['Massa Modal Teste'],
              labels=['Teste 01', 'Teste 02'],
              colors=color_map)

plt.title('Frequência vs Massa Modal', fontsize=20)

plt.xlabel('Frequência (Hz)', fontsize=15)
plt.ylabel('Massa Modal (kg)', fontsize=15)

plt.legend(loc='upper right')

ax = ax
```
